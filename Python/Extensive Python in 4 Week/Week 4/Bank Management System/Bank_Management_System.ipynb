```python
import pickle
import os
import pathlib

class BankAccount:
    accNo = 0;
    name = ''
    deposit=0
    accountType=''
    
    def createAccount(self):
        self.accNo = int(input("Please Enter the Account Number: "))
        self.name = input("Please Enter the name of the Account holder: ")
        self.accountType = input("Please Enter the type of the account [C/S]: ")
        self.deposit = int(input("Please Enter the initial deposit amount(>=500 for Savings and >=1000 for Current"))
        print("\n\n\nAccount has been successfully Created")
        
    def showAccount(self):
        print("Account Number: ", self.accNo)
        print("Account Holder Name: ", self.name)
        print("Type of Account:", self.accountType)
        print("Account Balance: ", self.deposit)
        
    def modifyAccount(self):
        print("Account Number: ", self.accNo)
        self.name = input("Account Holder Name to Modify: ")
        self.accountType = input("The type of the account to modify: ")
        self.deposit = int(input("Enter the new amount: "))
        
    def depositAmount(self, amount):
        self.deposit += amount
    
    def withdrawAmount(self, amount):
        self.deposit -= amount
        
    def statement(self):
        print(self.accNo, " ", self.name, " ", self.accountType, " ", self.deposit)
        
    def getAccountNo(self):
        return self.accNo
    
    def getAccountHolderName(self):
        return self.name
    
    def getAccountType(self):
        return self.accountType
    
    def getDeposit(self):
        return self.deposit
    
def Introduction():
    print("\t\t\t\t******************************")
    print("\t\t\t\tPython Bank Management System")
    print("\t\t\t\t******************************")
    
    print("\t\t\t\Brought to you by Olalekan Samuel")
    print("\t\t\t\From the Extensive Python Fundamentals")
    input()
    
def writeAccount():
    account = BankAccount()
    account.createAccount()
    writeAccountsToFile(account)
    
def displayAll():
    file = pathlib.Path('accounts.data')
    if file.exists():
        infile = open('accounts.data', 'rb')
        mylist = pickle.load(infile)
        for item in mylist:
            print(item.accNo, " ", item.name, " ", item.accountType, " ", item.deposit)
        infile.close()
    else:
        print("No Record found that can be displayed")
        
def displaySpecificAccount(num):
    file = pathlib.Path('accounts.data')
    if file.exists():
        infile = open('accounts.data', 'rb')
        mylist = pickle.load(infile)
        infile.close()
        found = False
        for item in mylist:
            if item.accNo == num:
                print("The balance in the account is = ", item.deposit)
                found = True
    else:
        print("No Record found that match the account number")
    if not found:
        print("No Existing Record found that match this account number")
        
def depositAndWithdraw(num1, num2):
    file = pathlib.Path('accounts.data')
    if file.exists():
        infile = open('accounts.data', 'rb')
        mylist = pickle.load(infile)
        infile.close()
        os.remove('accounts.data')
        for item in mylist:
            if item.accNo == num1:
                if num2 == 1:
                    amount = int(input("Please Enter the amount you will like to deposit: "))
                    item.deposit += amount
                    print("Your account has been successfully updated")
                elif num2 == 2:
                    amount = int(input("Please Enter the amount you will like to withdraw: "))
                    if amount <= item.deposit:
                        item.deposit -= amount
                    else:
                        print("You cannot withdraw an amount more than the balance in the account")
                        
    else:
        print("No Record found that match the account number")
    outfile = open('newaccounts.data', 'wb')
    pickle.dump(mylist, outfile)
    outfile.close()
    os.rename('newaccounts.data', 'accounts.data')
    
def deleteAccount(num):
    file = pathlib.Path('accounts.data')
    if file.exists():
        infile = open('accounts.data', 'rb')
        oldlist = pickle.load(infile)
        infile.close()
        newlist = []
        for item in oldlist:
            if item.accNo != num:
                newlist.append(item)
        os.remove('accounts.data')
        outfile = open('newaccounts.data', 'wb')
        pickle.dump(newlist, outfile)
        outfile.close()
        os.rename('newaccounts.data', 'accounts.data')

def modifyAccount(num):
    file = pathlib.Path('accounts.data')
    if file.exists():
        infile = open('accounts.data', 'rb')
        oldlist = pickle.load(infile)
        infile.close()
        os.remove('accounts.data')
        for item in oldlist:
            if item.accNo == num:
                item.name = input("Enter the name of the account holder you wish to modify")
                item.accountType = input("Enter the account type you wish to modify")
                item.deposit = int(input("Enter the new amount"))
        outfile = open('newaccounts.data', 'wb')
        pickle.dump(oldlist, outfile)
        outfile.close()
        os.rename('newaccounts.data', 'accounts.data')
    
def writeAccountsToFile(account):
    file = pathlib.Path("accounts.data")
    if file.exists():
        infile = open('accounts.data', 'rb')
        oldlist = pickle.load(infile)
        oldlist.append(account)
        infile.close()
        os.remove('accounts.data')
    else:
        oldlist = [account]
    outfile = open('newaccounts.data', 'wb')
    pickle.dump(oldlist, outfile)
    outfile.close()
    os.rename('newaccounts.data', 'accounts.data')

#Start the Bank Management System
choice = ''
num = 0
Introduction()

while choice != 8 :
    print("\tMAIN MENU")
    print("\t1. NEW ACCOUNT")
    print("\t2. DEPOSIT MONEY")
    print("\t3. WITHDRAW MONEY")
    print("\t4. BALANCE ENQUIRY")
    print("\t5. DISPLAY ALL ACCOUNT HOLDERS")
    print("\t6. CLOSE AN ACCOUNT")
    print("\t7. MODIFY AN ACCOUNT")
    print("\t8. EXIT")
    print("\tMake a Choice (1-8) ")
    #choice = input()
    
    if choice == '1':
        writeAccount()
        
    elif choice == '2':
        num = int(input("Please Enter the account number: "))
        depositAndWithdraw(num, 1)
    
    elif choice == '3':
        num = int(input("Please Enter the account number: "))
        depositAndWithdraw(num, 2)
        
    elif choice == '4':
        num = int(input("Please Enter the account number: "))
        displaySpecificAccount(num)
        
    elif choice == '5':
        displayAll()
    
    elif choice == '6':
        num = int(input("Please Enter the account number: "))
        deleteAccount(num)
    
    elif choice == '7':
        num = int(input("Please Enter the account number: "))
        modifyAccount(num)
    
    elif choice == '8':
        print = ("Thank you for using the system. Have a good day. Bye: ")
        break
        
    else:
        print("You have made an invalid choice")
    
    choice = input("Please Enter your choice: ")
```
